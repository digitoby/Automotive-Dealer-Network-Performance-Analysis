In this notebook, I created a synthetic dataset designed to simulate real-world dealership performance data across Kiaâ€™s U.S. regions. It models how operational, financial, and customer satisfaction metrics interact over time, allowing for exploratory, reporting, and predictive analysis relevant to a Regional Fixed Operations Analyst role.

Below, you will see code for generating the dataset and some EDA before throwing the data into Tableau.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Creating the dataset

```python
np.random.seed(42)

# --- Parameters ---
dealers = [f"D{str(i).zfill(3)}" for i in range(1, 51)]
regions = ["West", "Midwest", "South", "Northeast"]
months = pd.date_range(start="2023-01-01", periods=24, freq="MS")

# --- Regional performance profiles ---
region_profiles = {
    "West": {"service_mu": 250000, "service_sd": 20000, "csi_mu": 82, "csi_sd": 3, "retention_mu": 0.6},
    "Midwest": {"service_mu": 200000, "service_sd": 15000, "csi_mu": 87, "csi_sd": 3, "retention_mu": 0.7},
    "South": {"service_mu": 180000, "service_sd": 15000, "csi_mu": 90, "csi_sd": 2, "retention_mu": 0.8},
    "Northeast": {"service_mu": 260000, "service_sd": 25000, "csi_mu": 80, "csi_sd": 4, "retention_mu": 0.55},
}

# --- Generate data ---
data = []

for dealer in dealers:
    region = np.random.choice(regions)
    profile = region_profiles[region]
    
    dealer_factor = np.random.normal(1, 0.05)  # dealer-level offset
    
    for month in months:
        # Seasonal pattern (summer peaks)
        season_factor = 1 + 0.2 * np.sin((month.month - 1) / 12 * 2 * np.pi)
        
        service_revenue = np.random.normal(profile["service_mu"], profile["service_sd"]) * dealer_factor * season_factor
        parts_revenue = service_revenue * np.random.uniform(0.3, 0.5)
        
        repair_orders = np.random.normal(1500, 300) * season_factor
        avg_turnaround_days = np.random.normal(2.5, 0.5) + np.random.normal(0, 0.3) * (repair_orders / 2000)
        
        csi_score = np.random.normal(profile["csi_mu"], profile["csi_sd"]) - (avg_turnaround_days - 2.5) * 2
        retention_rate = np.clip(profile["retention_mu"] + (csi_score - 85) / 100, 0.3, 0.9)
        
        warranty_claims = np.random.poisson(25 - 5 * np.sin((month.month - 1) / 12 * 2 * np.pi))
        
        data.append([dealer, region, month, service_revenue, parts_revenue, repair_orders,
                     csi_score, retention_rate, avg_turnaround_days, warranty_claims])

df = pd.DataFrame(data, columns=[
    "dealer_id", "region", "month", "service_revenue", "parts_revenue",
    "repair_orders", "csi_score", "retention_rate", "avg_turnaround_days", "warranty_claims"
])

# Round numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].round(2)

df.head()
```
